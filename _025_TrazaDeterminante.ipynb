```python
v1 = np.array([0,1])
v2 = np.array([1,0])

graficarVectores([v1,v2],["blue","red"])
```

```python

A = np.array([[2,0],[0,2]])
print(A)
```

```python
v1_tra = A @ v1
v2_tra = A @ v2

graficarVectores([v1_tra,v2_tra],["blue","red"])
```

```python
det_A = np.linalg.det(A)
print(det_A)
```

```python
area_tran = abs((v1_tra[0]-v2_tra[0])) * abs((v1_tra[1]-v2_tra[1]))
print(area_tran)
```
