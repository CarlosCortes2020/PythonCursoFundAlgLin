```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5)
y_1 = 3*x + 5
y_2 =2*x + 3
```

```python
plt.figure(figsize=(10,6))
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5) 
plt.plot(x, y_1, label='y=3x+5')
plt.plot(x, y_2, label='y=2x+3')

#Encuentra el punto de interseccion
A = np.array([[3, -1],
              [2, -1]])
B = np.array([[ -5], 
              [ -3]])

sol = np.linalg.solve(A, B)
print(f'La solucion es: {sol.flatten()}')
plt.plot(sol[0], sol[1], 'ro')
plt.text(sol[0], sol[1], f'  {sol.flatten()}', fontsize=12, verticalalignment='bottom')
plt.legend()
plt.show()  
```
