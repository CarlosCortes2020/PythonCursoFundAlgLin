```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v1 = np.array([2, 7])
v2 = np.array([3, 5])
```

```python
v1v2 = v1 + v2
print(v1v2)
```

```python
np.linalg.norm(v1v2)
```

```python
norma_v1 = np.linalg.norm(v1)
norma_v2 = np.linalg.norm(v2)
sumv1v2 = norma_v1 + norma_v2
norma_v1v2 = np.linalg.norm(v1v2)

print(norma_v1)
print(norma_v2)
print(sumv1v2)
print(norma_v1v2)

print(norma_v1v2 <= norma_v1 + norma_v2)
```

```python
v1 = np.array([0,0,2,7])
v2 = np.array([0,0,3,5])

v1_aux = np.array([v1[2],v1[3],v2[2],v2[3]])
v1v2 = np.array([0,0,5,12])
```

```python
plt.quiver([v1[0], v1_aux[0], v1v2[0]],
           [v1[1], v1_aux[1], v1v2[1]],
           [v1[2], v1_aux[2], v1v2[2]],
           [v1[3], v1_aux[3], v1v2[3]],
           angles = 'xy', scale_units = 'xy', scale = 1,
           color = sns.color_palette())

plt.xlim(-0.5, 6)
plt.ylim(-0.5, 15)
```
